## 3.5 张量算法的基本性质

按元素的一元运算都不会改变其操作数的形状。给定具有相同形状的任意两个张量，任何按元素的二元运算的结果都不会改变其操作数的形状。

```python
A = torch.arange(20, dtype=torch.float32).reshape(5,4)
B = A.clone()
A, A+B
```

两个矩阵的按元素乘法称为Hadamard 积（数学符号$\odot$）

```python
A * B
```

将张量乘以或加上一个标量不会改变张量的形状，其中每个元素都将以标量相加或相乘

```python
a = 2

X = torch.arange(24).reshape(2, 3, 4)
a + X, (a * X).shape
```

## 3.10 范数
范数（norm）。表示一个向量有多大。
线性代数中，向量范数是将向量映射到标量的函数f。给定任意向量$\mathbf x$，向量范数要满足一些属性。

性质1：按常数因子$\alpha$，其范数也会按照相同的常数因子的绝对值缩放
$$
f(\alpha \mathbf x)=|\alpha|f(x)
$$
性质2：三角不等式
$$
f(\mathbf{x+y})\le f(x) + f(y)
$$
性质3：范数非负
$$
f(x)\ge 0
$$
性质4：范数最小为0，当且仅当向量由全0组成
$$
\forall i,[\mathbf x]_i=0\Leftrightarrow f(\mathbf x)=0
$$

 L2范数
 $$
||\mathbf x||_2=\sqrt{\sum_{i=1}^nx_i^2}
 $$

```python
u = torch.tensor([3.0,-4.0])
torch.norm(u)
```

L1范数，表示为向量元素的绝对值之和：
$$
||\mathbf x||_1=\sum_{i=1}^n|x_i|
$$

```python
torch.abs(u).sum()
```

Lp范数
$$
||\mathbf x||_p=\bigg(\sum_{i=1}^n|x_i|^p \bigg)^{1/p}
$$

矩阵的Frobrnius范数是矩阵元素平方和的平方根
$$
||X||_F=\sqrt{\sum_{i=1}^m\sum_{j=1}^n x_{ij}^2}
$$

```python
torch.norm(torch.ones((4,9)))
```
